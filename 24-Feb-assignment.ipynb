```python
Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.
Ans :
```

```python
df=pd.DataFrame({'A':['Corbett National Park','Bandipur National Park','Bannerghatta National Park','Gir National Park'],'B':[1936,1974,1974,1975],'C':['Uttarakhand','Karnataka','Karnataka','Gujarat']})
```

```python
index=list(df.index)
index
```

```python
def reindex_df(data):
    index=list(df.index)
    for i in range(len(index)):
        index[i]=index[i]+i+1
    return data.set_index([index])   
```

```python
reindex_df(df)
```

```python
df=pd.DataFrame({'Text':['Royal Challengers Bangalore','Sunrisers Hyderabad','Visakhapatnam','IPL-2017']})
df
```

```python
Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
median, and standard deviation of the values in the 'Values' column.
```

```python
df=pd.DataFrame({'Values':[10, 20, 30, 40, 50]})
```

```python
def prop_calc(data):
    mean=df['Values'].mean()
    median=df['Values'].median()
    std=df['Values'].std()
    return f"Mean: {mean}, Median: {median}, Standard Deviation: {round(std)}"
```

```python
prop_calc(df)
```

```python
Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.
```

```python
df=pd.DataFrame({'Sales':[110,120,210,655,210,200,310,230,411,420],'Date':pd.date_range(start='5/1/2023', end='5/10/2023')})
def movingAverage (data):
    data['MovingAverage']=data['Sales'].rolling(window=7,min_periods=1).mean()
    return data
movingAverage(df)
    
```

```python
Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
Ans:
    
```

```python
df=pd.DataFrame({'Date':['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05']})
def weekDays(data):
    data['Weekday']=pd.to_datetime(data['Date']).dt.day_name()
    return data
weekDays(df)
```

```python
Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.
```

```python
df=pd.DataFrame({'Date':['2023-01-05 10:10:10','2019-12-31 23:40:10','2023-01-12 13:10:10','2018-02-01 10:10:10','2023-01-16 04:40:10','2023-01-20 18:10:10']})
df
```

```python
def selectDate(data):
    return data[(pd.to_datetime(data['Date'])>=pd.to_datetime('2023-01-01')) & (pd.to_datetime(data['Date'])<=pd.to_datetime('2023-01-31'))]
selectDate(df)                                              
```
