```python
Q2. Give advantages and disadvantages of using API.
Ans :
    Advatages :
        1. API can be easily integrated into any application irrespective of the platform or technology used.
        2. With API, developers can build complex application without worrying about code that goes behind it.
        3. API helps developers and companies in automating the process to enhance their productivity and 
           performance.
        4. APIs are highly customizable and adaptable thus allowing us to use right to update and implement 
           them in the way they desire.
    
    Disadvantages :
        1. Creating API is very time-consuming process.
        2. Since API are single point of entry, they are vulnerable to attack by hackers.
        3. Once API is compromised, it leads to loss of other related or attached system.
        4. APIs are difficult to manage.
        5. There are chances of system crashing during API testing.
```

```python
Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.
Ans :
    REST:
        
    REST (Respresentational State Transfer) is a software architectural style that defines the set of rules 
    used for creating web services. 
    Web services which follow REST architectural style are known as RESTful web services.
    It allows requesting systems to access and manipulate web resources by using a uniform and predefined set 
    of rules.
    Request for data sent to REST API is usually done through HTTP. Once request is received, API designed for 
    REST ( called RESTful API) can return message in variety of formats : HTML, XML, plain text and JSON. JSON 
    is favoured as message format because it can be read by any programming langugae and it is lightweight.
    
    RESTful Services are based on the following principles:
        1. Client-server architecture: it consists of client, server and resources.
        2. Stateless client-server communication : no client content is stored on 
           the server between requests. Information about the session's state is 
           instead held with the client.
        3. Cacheable data to eliminate the need for some client-server interaction.
        4. Uniform interface : RESTful services use a uniform interface consist of 
           resources, HTTP verbs (GET, POST, PUT, DELETE) and hypermedia links.
        5. Layered system: RESTful services can be composed of multiple layers, allowing for scalability, flexibility and security..
        
        SOAP:
            
            SOAP (Simple object access protocol) is a network protocol for exchanging
            structured data between nodes. SOAP can be carried over a variety of standard 
            protocols including HTTP.
            
            SOAP is a lightweight protocol used to create web APIs usually with XML. It 
            supports wide range of communication protocol across the internet like HTTP,
            SMTP and TCP.
            SOAP messages are XML documents that are comprised of the following three basic
            building blocks:
                1. SOAP envelope that encapsulates all the data in the message and identify
                   XML document as SOAP message.
                2. Header element contains additional information about the SOAP message.
                3. The Body element includes the details of the actual message that need
                   to be sent from the web service to the calling application. This data includes 
                   reponses information.
    
    Shortcomings of SOAP :
        1. SOAP used only XML format data in web services, whearas JSON and other light-weight formats 
           are not supported by it.
        2. It is slow because it uses XML format, whereas payload for simple string message is large.
        3. SOAP is significantly more complex than REST
        4. There is no state reference for the remote object in the SOAP client.
        
        
```
