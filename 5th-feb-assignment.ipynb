```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
Ans:
    class - blueprint for the object. Classes contain methods (functions) and attributes.
            Using classes, you can create multiple objects with the same behavior instead of 
            writing their code multiple times.
    object - a real world entity. objects are constructed from class blueprint that contain 
             method and attributes.
ex. in the example below, Employee is a class with attributes first, last, age and salary and method 
    display_salary while akshay and vijay are the projects.

class Employee:
  def __init__(self,first,last,age,salary):
    self.first=first
    self.last=last
    self.age=age
    self.salary=salary
  
  def display_salary(self):
    return f"The salary of {self.first} {self.last} is {self.salary}"

akshay=Employee('Akshay','Jha',21,25000)
vijay=Employee('Vijay','Khurana',25,50000)
```

```python
Q3. Explain why the __init__() function is used. Give a suitable example.
Ans :
    __init__() is a special method in classes which gets called every time 
    when you create an object of the class (instantiate). 
    This type of function is called constructor function.
    The __init__ method only allows the class to initialize the object's attributes. 
    It is only used within classes.
```

```python
Q5. What is inheritance? Give an example for each type of inheritance.
Ans :
    It is the ability by which parent class inherit their properties to the child class.
    It promotes the code reusability.
    
    Types of inheritance :
        1. Single inheritance : here single child is derived from single parent.
        
        ex. : in the below example the class Employee is inherited from parent class Human
        
            class Human:
                def __init__(self,first,last,age):
                    self.first=first
                    self.last=last
                    self.age=age
    
                def full_name(self):
                    return f"{self.first} {self.last}"

            class Employee(Human):
                def __init__(self,first,last,age,salary):
                    super().__init__(first,last,age)
                    self.salary=salary

            sanjay=Employee('Sanjay','Singhania',32,100000)
            sanjay.full_name()
        
        2. Multiple inheritance : here single child is inherited from two or more parent class
        
        ex. in the example below class Penguine is inherited from parent class Aquatic and Ambulatory
           
            class Ambulatory:
                    def __init__(self,name):
                        self.name=name

                    def walk(self):
                        return f"{self.name} is walking"

            class Aquatic:
                def __init__(self,name):
                    self.name=name

                def swim(self):
                    return f"{self.name} is swimming"

            class Penguine(Ambulatory,Aquatic):
                def __init__(self,name):
                    super().__init__(name=name)

            captain_cook=Penguine('Captain Cook')
            captain_cook.swim()
            captain_cook.walk()
            
            
        3. Multilevel inheritance : When child class is inherited from parent class which is inherited 
                                    from another parent class. Such type of inheritance is called 
                                    multilevel inheritance.
            
            ex. in the example below class Son is inherited from Father which is inherited from Grandfather
            
            class Grandfather:
                def __init__(self,grandfather_name):
                    self.grandfather_name=grandfather_name

            class Father(Grandfather):
                def __init__(self,grandfather_name,father_name):
                    super().__init__(grandfather_name)
                    self.father_name=father_name

            class Son(Father):
                def __init__(self,grandfather_name,father_name,name):
                    super().__init__(grandfather_name,father_name)
                    self.name=name

                def full_name(self):
                    return f"{self.name} {self.father_name} {self.grandfather_name}"

            aniket=Son('Rishabh','Rajesh','Aniket')
            aniket.full_name()
            
        4. Hierarchical Inheritance : There are multiple child class which is inherited from same parent.
        
        ex. in the example below child class Employee and Doctor is inherited from Details.
        
                class Details:
                    def __init__(self,id,name,gender):
                        self.id=id
                        self.name=name
                        self.gender=gender
                    def show_data(self):
                        print(f"ID: {self.id}    Name : {self.name}   Gender : {self.gender}")

                class Employee(Details):
                    def __init__(self,id,name,gender,company,dept):
                        super().__init__(id,name,gender)
                        self.company=company
                        self.dept=dept

                    def show_employee(self):
                        self.show_data()
                        print(f"Company : {self.company}       Department : {self.dept}")

                class Doctor(Details):
                    def __init__(self,id,name,gender,hospital,dept):
                        super().__init__(id,name,gender)
                        self.hospital=hospital
                        self.dept=dept

                    def show_employee(self):
                        self.show_data()
                        print(f"Hospital : {self.hospital}       Department : {self.dept}") 

                ajay=Employee(112345,'Ajay','Male','TCS','Web Developer')
                nisha=Doctor(58798579,'Nisha','Female','AIIMS','Neurologist')
            
        
        5. Hybrid Inheritance : It is a mixture of more than one type of inheritance. 
        ex. in the example below child class Employee is derived from two different type of inheritance class
        1) Human - direct class 2) Student - single inhritance class
        
        class Human: 
            def __init__(self,first,last,age):
                self.first = first
                self.last = last
                self.age = age

            def details(self):
                print(f'Name : {self.first} {self.last}')
                print(f'Age : {self.age}')
                
        class Student(Human):

            def __init__(self,first,last,age,std):
                super().__init__(first,last,age)
                self.std=std

            def details(self):
                print(f'Name : {self.first} {self.last}')
                print(f'Age : {self.age}')
                print(f'Class : {self.std}')
        

        class Employee(Student,Human):
            def __init__(self,first,last,age,std,job):
                super().__init__(first,last,age,std)
                self.job=job

            def details(self):
                print(f'Name : {self.first} {self.last}')
                print(f'Age : {self.age}')
                print(f'Class : {self.std}')
                print(f'Class : {self.job}')

        sumit=Student('Sumit','Raina',18,'XII')
        amit=Employee('Amit','Poghat',21,'B.Tech','Web Developer')

sumit.details()
amit.details()
          
        
                
            
        
        

```
