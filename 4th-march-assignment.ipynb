```python
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.
```

```python
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.
```

```python
import plotly.express as px
iris = px.data.iris()
iris.head()
```

```python
fig = px.scatter_matrix(iris,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],color='species')
fig.show()
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans :
Distplots are used for univariant analysis on a continuous or numerical data. Displots are the combination of 3 graphs :

Histogram
KDE(Kernal Density Estimation) plot
RugPlot

Distplots are used for univariant analysis on a continuous or numerical data.
Displots are the combination of 3 graphs :
1. Histogram
2. KDE(Kernal Density Estimation) plot
3. RugPlot
```

```python
import plotly.figure_factory as ff
import seaborn as sns
titanic= sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
hist_data=[titanic[titanic['survived']==1]['age'],titanic[titanic['survived']==0]['age']]
group_labels = ["Survived", "Not Survived"]
fig = ff.create_distplot(hist_data,group_labels)
fig.show()
```
