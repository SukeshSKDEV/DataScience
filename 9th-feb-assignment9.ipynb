```python
Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.

class vehicle:
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.name_of_vehicle=name_of_vehicle
        self.max_speed=max_speed
        self.average_of_vehicle=average_of_vehicle
```

```python
Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.

class car(vehicle):
    def seating_capacity(self,capacity):
        return self.name_of_vehicle,capacity
bmw=car('BMW',250,150)
print(bmw.seating_capacity(8))
```

```python
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.
Ans: When child class is derived from two or more parent class it is called multiple inheritance.

ex. In the following code, we have inherited Penguine child classes from the two parent subclasses Aquatic 
    and Ambulatory. Penguine inheriting methods swim and walk from both the parent class. Thus showing
    multiple inheritance.

class Aquatic:
    def __init__(self,name):
        self.name=name
        
    def swim(self):
        return f"{self.name} is swimming"
    
    
class Ambulatory:
    def __init__(self,name):
        self.name=name

    def walk(self):
        return f"{self.name} is walking"
    

    
class Penguine(Aquatic,Ambulatory):
   pass

captain_cook=Penguine("Captain Cook")
print(captain_cook.swim())
print(captain_cook.walk())
```

```python
Q5.What is method overriding in python? Write a python code to demonstrate method overriding.
Ans: Method overriding is one of the concept in Polymorphism which allow us change the implementation 
of function of child class which was derived from the parent class.
It is the ability of a child class to change the implementation of any method which is already provided 
by one of its parent class.

ex. in the below example class Animal with method sound will be override by its child class' sound method.
Both child class are inherited from Animal class.

class Animal:
    def sound(self):
        return "wow wow"

class Cat(Animal):
    def sound(self):
        return "meow meow"

class Dog(Animal):
    def sound(self):
        return "woof woof"
    
cat=Cat()
cat.sound()
dog=Dog()
dog.sound()
```
