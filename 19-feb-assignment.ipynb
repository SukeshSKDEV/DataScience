```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
Ans : Fask is a python-based microweb framework. It is enable developers to create and scale web apps quickly
      and simply.
      It has a small and easy-to-extend core: it’s a microframework that doesn’t include an ORM (Object 
      Relational Manager) or such features.
      It does have many cool features like url routing, template engine. 
    
    Advantages of Flask Framework :
        1. Scalable: If you want to make an app that starts small, but has the potential to grow quickly 
           and in directions you haven't completely worked out yet, then falsk is an ideal choice.
        2. Easy to understand and lightweight.
        3. Integrated testing tools
        4. Built-in development server and fast debugger
        5. It is easy to deploy the flask in production
        6. RESTful request handling
```

```python
Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.
Ans :
    code :
        from flask import Flask

        app = Flask(__name__)

        @app.route("/")
        def hello_world():
            return "<h1>Hello World!!</h1>"

        if __name__=="__main__":
            app.run(host="0.0.0.0")
```

```python
Q3. What is App routing in Flask? Why do we use app routes?
Ans : 
    App routing is process of mapping specific URL with associated function that is intended to perform 
    specific task.
    We use app routes to access specific page of web app.
    
    ex:
    We can use app routes in the following way:
        In the below app, "/" route denotes root url. let suppose our site domain is www.sukesh.com,
        if we want to add routing www.sukesh.com/about, then we can use "/about"
    
    
    from flask import Flask

        app = Flask(__name__)

        @app.route("/")
        def hello_world():
            return "<h1>Hello World!!</h1>"
        
        @app.route("/about")
        def about_us():
            return "<h1>About Us</h1>"

        if __name__=="__main__":
            app.run(host="0.0.0.0")
    
```

```python
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

Ans :
    
    from flask import Flask

    app = Flask(__name__)

    @app.route("/")
    def company():
        return "<h1>Company Name: ABC Corporation <br> Location: India <br> Contact Detail: 999-999-9999 </h1>"

    @app.route("/welcome")
    def welcome():
        return "<h1>Welcome to ABC Corporation</h1>"

    if __name__=="__main__":
        app.run(host="0.0.0.0")
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
Ans :
    url_for() is used in Flask for URL Building.
    
    ex. In the code below, if user type is admin, then it is redirected to the "/admin" page while in other case
        it is redirected to the "/user" page.
    
    from flask import Flask, redirect, url_for

    app = Flask(__name__)

    @app.route("/admin")
    def admin():
        return "You logged in the admin page."

    @app.route("/user")
    def user():
        return "You logged in the user page"

    @app.route("/user_type/<person>")
    def type_of_user(person):
        user_type=""
        if person=="admin":
            user_type="admin"
        else:
            user_type="user"

        return redirect(url_for(user_type))



    if __name__=="__main__":
        app.run(host="0.0.0.0")
```
