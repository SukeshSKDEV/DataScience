```python
Q1. Explain GET and POST methods.
Ans :
    GET : It is used to retrive information.
          Data is sent via query string.
          Information is plainly visible in the URL.
          Limited amount of data can be sent.
    POST : Used to post data to the server.
           Used to write/create/update
           Data is sent via request body, not a query string.
           Can send any sort of data ex. JSON
           POST is used to send sensitive information like password.
```

```python
Q2. Why is request used in Flask?
Ans :
    To recover all the data sent by the client to the server, request is used in flask.
    request object contains all data the sent by the clinet to the server.
    This data can be recovered using HTTP request methods like GET and POST.
```

```python
Q3. Why is redirect() used in Flask?
Ans: 
    redirect() function is use to redirect the user to different location/url.
    
    syntax: redirect(location,status code)
            location=url where you want to redirect user
            status code=the default redirect code is 302 which means user is redirected tempororily
                        There are some other status code like 300, 301, 302, 303, 304,305, 306, 307
        
        ex. In the  below code, if page_location is home then it redirects to '/home', if page_location is about
            then it redirects to '/about' page.
        
            from flask import Flask, redirect, url_for
            app = Flask(__name__)


            @app.route('/home')
            def home_page():
                return 'Welcome to the Home page'


            @app.route('/about')
            def about():
                return "This site contains tutorial on Flask"


            @app.route('/page_location/<string:page>')
            def page_location(page):
                location=""
                if page == 'home':
                    location = 'home'

                if page == 'about':
                    location = 'about'
                return redirect(url_for(location))


            if __name__ == '__main__':
            app.run(debug=True)    
```

```python
Q4. What are templates in Flask? Why is the render_template() function used?
Ans :
    Templates are the documents that contain static data and placeholder for dynamic data. A template is render
    with specific data to produce final document.
    Flask uses jinja template library to render template.
    We can use {{ }} expression to display dynamic data and <% %> expression to denote control flow like if and for.
    
    render_template() function is used to generate output from templates based on Jinja2 engine. 
    templates are present in the 'templates' folder.    
```
