```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans : 
    1. Line plot (sns.lineplot) :
        Uses : It is used to show the relationship between continuous or categorical values in a continuous 
               data point format. It helps to predict the outcome and analyse the variations in the data.
            
    2. Bar plot (sns.barplot) :
        Uses : It shows the relationship between a numeric and a categorical data.
               The height of each bar represents the value of a data point
            
    3. Histogram (sns.histplot) :
        Uses : A histogram provides a visual representation of the distribution of a dataset.
               It also helps to visualize whether the distribution is symmetric or skewed left or right. 
        
    4. Scatterplot (sns.scatterplot) :
        Uses : It is used to show the visual relationship between the two numeric variables. The values
               of the variables are represented by dots.
               The positioning of the dots on the vertical and horizontal axis will inform the value of 
               the respective data point
        
    5. Heatmap plot (sns.heatmap) :
        Uses : It visualizes the data in a 2-dimensional format in the form of colored maps.
               Color variation gives visual cues to the readers about the magnitude of numeric values.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

Ans : 
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(data=fmri,x='timepoint',y='signal',hue='event')
plt.title('Signal over time for different events')
plt.show()
```

```python
sns.lineplot(data=fmri,x='timepoint',y='signal',hue='region')
plt.title('Signal over time for different region')
plt.show()
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
df=sns.load_dataset('flights')
df.head()
```

```python
flights=pd.pivot(df,index=['year'],columns=['month'],values='passengers')
flights.head()
```

```python
corr=flights.corr()
corr.head()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr,cmap='Blues',annot=True,linewidths=.5)
plt.show()
```
