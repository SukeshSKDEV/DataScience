1. For loop is used when the number of iteration is known.
   While loop is used to run block of code again and again when the certain condition is 
   True, if the condition is false then loop breaks.
   
   l=['a','b','c','d','e','f','g','h']
   
   for i in l:
       print(i)
   
   Here, for loop will go in each element of a list and print it.
   
   l=['a','b','c','d','e','f','g','h']
   n=0
   while n<len(l) :
       print(l[n])
       n=n+1
    
   Here, i while loop untill and unless condition is True means value of n is less than 
   length of array l (i.e.8), loop will run and print element. When n become equal to or 
   greater than length of l, condition become False and loop break.

num_units=int(input('Enter the number of units of your electricity  bill per month :'))
total_price=0
temp=0

while num_units>0:
    if num_units>300 :
        temp=num_units-300        # 310-300=10*20=200 Rs
        total_price=total_price+temp*20
        num_units=num_units-temp
    elif num_units<=300 and num_units>200 :
         temp=num_units-200        # 300-200=10*10=100 Rs
         total_price=total_price+temp*10
         num_units=num_units-temp
    elif num_units<=200 and num_units>100 :
         temp=num_units-100        # 200-100=10*6=60 Rs
         total_price=total_price+temp*6
         num_units=num_units-temp
    else :
        total_price=total_price+num_units*4.5
        num_units=num_units-100

print('Electricity Bill : ',total_price)
